```python
import re
import pandas as pd
import openpyxl
```

```python
def extract_info_from_email(email_text):
    # Define regular expressions for each type of information
    patterns = {
        'First Name': r'First Name\s+(.*)',
        'Last Name': r'Last Name\s+(.*)',
        'Email Address': r'Email Address\s+(.*)',
        'Date of Birth': r'Date of Birth\s+(.*)',
        'Country': r'Country\s+(.*)',
        'Province/State': r'Province/State\s+(.*)',
        'City/Town': r'City/Town\s+(.*)',
        'Address': r'Address\s+(.*)',
        'Postal Code': r'Postal Code\s+(.*)',
        'Phone': r'Phone\s+(.*)',
        'Expected Start Date': r'Expected Start Date\s+(.*)',
        'Program(s) of Interest': r'Program\(s\) of Interest\s+(.*)',
        'High School': r'High School\s+(.*)',
        'Level of Interest': r'Level of Interest\s+(.*)',
        'Source': r'Source\s+(.*)',
        'Level of Study Interest': r'Level of Study Interest\s+(.*)'
    }

    # Initialize an empty dictionary to store the extracted data
    extracted_data = {}

    # Iterate over each pattern and extract the relevant information
    for key, pattern in patterns.items():
        match = re.search(pattern, email_text)
        if match:
            extracted_data[key] = match.group(1).strip()

    return extracted_data
```
